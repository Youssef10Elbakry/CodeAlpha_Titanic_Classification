```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing

from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from catboost.datasets import titanic
```

```python
titanic_train, titanic_test = titanic()
titanic_train.head()
```

```python
titanic_train.info()
```

```python
titanic_test.info()
```

```python
titanic_train.head()
```

```python
titanic_train.isna().sum()
```

```python
titanic_train.Age[titanic_train.Age.isna()] = titanic_train.Age.mean()
print(titanic_train.Age.isna().sum())
```

```python
titanic_train.drop('Cabin', axis = 1, inplace = True)
```

```python
titanic_train.head()
```

```python
titanic_train.Embarked[titanic_train.Embarked.isna()] = titanic_train.Embarked.mode()[0]
print(titanic_train.Embarked.isna().sum())
```

```python
X = titanic_train.drop('Survived',axis=1)
Y = titanic_train['Survived']
```

```python
x_train, x_test, y_train,y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
x_train.head()
```

```python
x_train = x_train.drop(['PassengerId','Name'],axis=1)
x_test = x_test.drop(['PassengerId','Name'],axis=1)
```

```python
x_train.head()
```

```python
import category_encoders as ce
```

```python
categoryStrFeatures = [col for col in x_train.columns if x_train[col].dtypes == 'O']
categoryStrFeatures
```

```python
enc = ce.OneHotEncoder(cols = categoryStrFeatures)
encoded_x_train= enc.fit_transform(x_train[categoryStrFeatures])
encoded_x_test = enc.transform(x_test[categoryStrFeatures])
```

```python
encoded_x_train.head()
```

```python
scaler = MinMaxScaler()
scaled_encoded_x_train = scaler.fit_transform(encoded_x_train)
scaled_encoded_x_test = scaler.transform(encoded_x_test)
```

```python

GNB = MultinomialNB()

GNB.fit(scaled_encoded_x_train, y_train)
```

```python
print("Training Accuracy",GNB.score(encoded_x_train,y_train))
```

```python
Ypredict = GNB.predict(scaled_encoded_x_test)
accuracy_score(y_test,Ypredict)
```
